```python
import numpy as np
import pandas as pd
from scipy import stats
```

### **One Trial (Bernauli Distribution)**

```python
p = 0.1
X = [0,1]
probs = [0.9,0.1]
```

```python
import seaborn as sns

sns.barplot(y=probs, x=X)
```

### **Two Trials**

```python
X = [0,1,2]
p = 0.1
probs = [
    #FF
    (1-p)*(1-p),
    #FS + SF
    (1-p)*p * 2,
    #SS
    p*p
]

sns.barplot(x=X, y=probs)
```

### **Three Trials**

```python
X = [0,1,2,3]
p = 0.1

probs = [
    #SSS
    p*p*p,
    #FSS SFS SSF
    (1-p)*p*p * 3,
    #SFF,FSF,FFS
    p*(1-p)*(1-p) * 3,
    #FFF
    (1-p)*(1-p)*(1-p)
]

sns.barplot(y=probs, x=X)
```

### **Binom PMF**

```python
from scipy.stats import binom

X = np.arange(0,10,1)
p = 0.1
binom_probs = binom.pmf(X,n=10, p=p)

sns.barplot(x=X, y=binom_probs)
```

**BINOM CDF**

```python
X = np.arange(0,10,1)
p = 0.1
binom_probs = binom.cdf(X,n=10, p=p)
sns.barplot(x=X, y=binom_probs)
```
