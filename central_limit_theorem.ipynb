```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weight-height.csv')['Height']
```

```python
df.head()
```

```python
import seaborn as sns

sns.histplot(x = df, kde = True)
```

```python
len(df)
```

```python
mean_pop, std_pop = df.mean(), df.std()
mean_pop, std_pop
```

```python
# Sample

df.sample(5).describe()
```

```python
# Central Limit Theorem

samples_disb = [np.mean(df.sample(5)) for _ in range(20000)]
```

```python
sns.histplot(samples_disb)
```

```python
np.mean(samples_disb), np.std(samples_disb)
```

```python
# Central Limit Theorem

samples_disb = [np.mean(df.sample(25)) for _ in range(20000)]
```

```python
sns.histplot(samples_disb)

np.mean(samples_disb), np.std(samples_disb)
```

```python
std_pop
```

```python
n = 25
std_err = std_pop/np.sqrt(n)
std_err
```

```python
sns.histplot(samples_disb, bins = 1000)
```
